```python
import numpy as np
import pandas as pd

movies_name = pd.read_csv('movies.csv')
ratings_data = pd.read_csv('ratings.csv')

movies_name.head()


```

```python
ratings_data.head()
```

```python
movie_data = pd.merge(ratings_data, movies_name, on='movieId')
movie_data.head()
```

### Popularity Based Recommendations
Based on Highest Ratings and Number of Views


```python
movie_data.groupby('title')['rating'].mean().head()
```

```python
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())
ratings_mean_count.head()
```

```python
ratings_mean_count['rating'] = round(ratings_mean_count['rating'], 1)
ratings_mean_count.head()
```

```python
ratings_mean_count = ratings_mean_count[(ratings_mean_count['rating_counts'] > 100) & (ratings_mean_count['rating'] > 3)]
ratings_mean_count.head()
```

```python
ratings_mean_count = ratings_mean_count.sort_values('rating', ascending=False).head(10)
ratings_mean_count
```
